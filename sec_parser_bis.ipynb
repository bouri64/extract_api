```python
import re
def find_context(doc):    
    before = 2000
    after = 800
    pattern = "(?:Basic earnings|Basic net income|Earnings|Net income) per (?:common )?share"
    first_match_data = None
    # regex = re.compile(pattern, re.IGNORECASE)
    regex = re.compile(pattern)


    for page_number, page in enumerate(doc):
        text = page.get_text()
        matches = list(regex.finditer(text))

        if matches and first_match_data is None:
            match = matches[0]
            start, end = match.span()
            context = text[max(0, start - before): min(len(text)-1, end + after)]
            return context
    return None
```
