## Yfinance

```python
# Only price, volume and dividend
```

```python
import yfinance as yf
import pandas as pd

all_results = []

fiscal_years = range(2011, 2026)  # FY2011–FY2025

for symbol in companies:
    ticker = yf.Ticker(symbol)

    prices = ticker.history(start="2009-01-01", auto_adjust=False)
    dividends = ticker.dividends

    # Normalize timezones (critical)
    prices.index = prices.index.tz_localize(None)
    dividends.index = dividends.index.tz_localize(None)

    for fy in fiscal_years:
        fiscal_end = pd.Timestamp(f"{fy}-09-30")

        # Last trading day on or before Sep 30
        price_row = prices.loc[:fiscal_end].iloc[-1]

        close_price = price_row["Close"]
        adj_close = price_row["Adj Close"]
        volume = price_row["Volume"]
        close_date = price_row.name.date()

        # Dividend window: Oct 1 → Sep 30
        div_start = pd.Timestamp(f"{fy-1}-10-01")
        div_end = fiscal_end

        yearly_dividend = dividends.loc[div_start:div_end].sum()

        all_results.append({
            "Ticker": symbol,
            "Fiscal Year": fy,
            "Close Date": close_date,
            "Close Price": round(close_price, 2),
            "Adj Close": round(adj_close, 2),
            "Volume": int(volume),
            "Dividends Paid": round(yearly_dividend, 4)
        })

# Final dataframe
df = pd.DataFrame(all_results)
df.to_csv("./data/yfinance.csv", index=False)
```

```python
def find_value(items, label):
    """
    items: list of dicts like [{'label': ..., 'value': ...}, ...]
    label: the label to find
    """
    if not items:
        return None
    for item in items:
        # Sometimes label can be 'eps', sometimes 'EPS'
        if item.get("label", "").lower() == label.lower():
            return item.get("value")
    return None
```

```python
import pandas as pd
import finnhub

finnhub_client = finnhub.Client(api_key=API_KEY)

def find_value(items, label):
    if not items:
        return None
    for item in items:
        if item.get("label", "").lower() == label.lower():
            return item.get("value")
    return None

rows = []

for symbol in companies:
    fin = finnhub_client.financials_reported(symbol=symbol, freq='yearly')
    
    for year_data in fin['data']:
        end_date = year_data.get('endDate')
        if not end_date:
            continue
        year = int(end_date[:4])
        
        report = year_data['report']
        bs = report.get('bs', [])
        ic = report.get('ic', [])
        
        # Extract metrics safely
        eps_basic = find_value(ic, "Basic (in dollars per share)")
        eps_diluted = find_value(ic, "Diluted (in dollars per share)")
        current_assets = find_value(bs, "Total current assets")
        assets = find_value(bs, "Total assets")
        current_liabilities = find_value(bs, "Total current liabilities")
        liabilities = find_value(bs, "Total liabilities")
        debt = 0
        debt = find_value(bs, "Term debt") # redundant
        equity = find_value(bs, "Total shareholders’ equity")
        
        rows.append({
            "Symbol": symbol,
            "Year": year,
            "EndDate": end_date,
            "EPS Basic": eps_basic,
            "EPS Diluted": eps_diluted,
            "Current Assets": current_assets,
            "Assets": assets,
            "Current Liability": current_liabilities,
            "Liabilities": liabilities,
            "Debt": debt,
            "Equity": equity
        })

df = pd.DataFrame(rows)
df = df.sort_values(["Symbol", "Year"]).reset_index(drop=True)
print(df.head())
```

```python
import pandas as pd

# Placeholder for all rows
rows = []

for company in companies:
    # Fetch data
    finances = finnhub_client.company_basic_financials(company, 'all')
    
    # Access annual series
    annual_series = finances.get('series', {}).get('annual', {})
    
    # Fetch each metric's data
    eps_data = annual_series.get('eps', [])
    pe_data = annual_series.get('pe', [])
    ev_data = annual_series.get('ev', [])
    
    # Build a dictionary keyed by period to merge metrics
    data_by_period = {}
    
    for entry in eps_data:
        period = entry['period']
        year = int(period.split("-")[0])
        data_by_period[period] = {"Symbol": company, "Year": year, "Period": period, "eps": entry['v']}
        
    for entry in pe_data:
        period = entry['period']
        if period in data_by_period:
            data_by_period[period]["pe"] = entry['v']
        else:
            year = int(period.split("-")[0])
            data_by_period[period] = {"Symbol": company, "Year": year, "Period": period, "pe": entry['v']}
            
    for entry in ev_data:
        period = entry['period']
        if period in data_by_period:
            data_by_period[period]["ev"] = entry['v']
        else:
            year = int(period.split("-")[0])
            data_by_period[period] = {"Symbol": company, "Year": year, "Period": period, "ev": entry['v']}
    
    # Append all periods for this company
    rows.extend(data_by_period.values())

# Convert to DataFrame
df = pd.DataFrame(rows)

# Optional: sort nicely
df = df.sort_values(["Symbol", "Year"]).reset_index(drop=True)

print(df.head())
```

```python
income = toolkit.get_income_statement()

eps_dict = {}
eps_diluted_dict = {}

for t in companies:
    eps_dict[t] = income.loc[(t, "EPS")].T
    eps_diluted_dict[t] = income.loc[(t, "EPS Diluted")].T
```

```python
balance = toolkit.get_balance_sheet_statement()

# Dictionaries to store per-ticker data
assets_dict = {}
liabilities_dict = {}
debt_dict = {}
equity_dict = {}
common_stock_dict = {}

for t in companies:
    assets_dict[t] = balance.loc[(t, "Total Assets")].T
    liabilities_dict[t] = balance.loc[(t, "Total Liabilities")].T
    debt_dict[t] = balance.loc[(t, "Total Debt")].T
    equity_dict[t] = balance.loc[(t, "Total Equity")].T
    common_stock_dict[t] = balance.loc[(t, "Common Stock")].T
```

```python
cashflow = toolkit.get_cash_flow_statement()

dividends_dict = {}

for t in companies:
    # Extract "Dividends Paid" for each ticker
    dividends_dict[t] = cashflow.loc[(t, "Dividends Paid")].T
    
    # Make values positive (optional; usually Cash Flow shows negative outflow)
    dividends_dict[t] = dividends_dict[t].abs()
```

```python
all_tickers_df = []

for t in companies:
    # Convert all Series to single-column DataFrames with proper column names
    ticker_df = pd.DataFrame({
        "Assets": assets_dict[t],
        "Liabilities": liabilities_dict[t],
        "Debt": debt_dict[t],
        "Equity": equity_dict[t],
        "EPS Basic": eps_dict[t],
        "EPS Diluted": eps_diluted_dict[t],
        "Dividends Paid": dividends_dict[t],
        "Common Stock": common_stock_dict[t]
    })
    
    # Add Ticker column
    ticker_df["Ticker"] = t
    
    # Reset index to make fiscal year a column
    ticker_df = ticker_df.reset_index().rename(columns={"index": "Fiscal Year"})
    
    all_tickers_df.append(ticker_df)

# Combine all tickers
final_df = pd.concat(all_tickers_df, ignore_index=True)

# Reorder columns
final_df = final_df[[
    "Ticker", "Fiscal Year", "Assets", "Liabilities", "Debt", "Equity",
    "EPS Basic", "EPS Diluted", "Dividends Paid", "Common Stock"
]]

final_df.to_csv("./data/financialToolKit.csv")
```
